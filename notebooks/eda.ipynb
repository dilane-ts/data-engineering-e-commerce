```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
customers = pd.read_csv("../dags/data/raw/df_Customers.csv")
print(customers.head())
print(customers.info())
```

```python
order_items = pd.read_csv("../dags/data/raw/df_OrderItems.csv")
print(order_items.head())
```

```python
orders = pd.read_csv("../dags/data/raw/df_Orders.csv", parse_dates=["order_delivered_timestamp", "order_estimated_delivery_date", "order_approved_at"])
print(orders.head())
```

```python
new_orders = order_items.merge(orders, on='order_id', how='inner')
new_orders["revenue"] = new_orders["price"] + new_orders["shipping_charges"]
new_orders["revenue"].head()
```

```python
import numpy as np
payments = pd.read_csv("../dags/data/raw/df_Payments.csv")
print(np.sum(payments["order_id"].value_counts() == 1), payments.shape)
payments.head(n=20)
```

```python
products = pd.read_csv("../dags/data/raw/df_Products.csv")
products.head()
```

```python
# orders.loc[orders["order_approved_at"] != None, :]
orders["order_status"].value_counts().plot(kind="bar")
```

```python
new_orders = order_items.merge(orders, on='order_id', how='inner') 
dim_facts = new_orders.loc[:, ["order_id", "product_id", "customer_id"]]

# delivery delay
dim_facts["delay_delivery"] = orders["order_estimated_delivery_date"] - orders["order_delivered_timestamp"]
dim_facts['delay_delivery'] = dim_facts["delay_delivery"].dt.days

# revenue 
dim_facts["revenue"] = new_orders["price"] + new_orders["shipping_charges"]

# payment method
new_orders = new_orders.merge(payments, on='order_id', how='inner')
dim_facts["payment_method"] = new_orders["payment_type"]

# date dim
dim_date = new_orders.loc[:, ["order_approved_at"]]
dim_facts["date_id"] = dim_date["order_approved_at"]

dim_date = dim_date.rename(columns={"order_approved_at": "date_id"})
dim_date["day"] = dim_date["date_id"].dt.day.fillna(-1).astype("int")
dim_date["month"] = dim_date["date_id"].dt.month.fillna(-1).astype("int")
dim_date["year"] = dim_date["date_id"].dt.year.fillna(-1).astype("int")
```
